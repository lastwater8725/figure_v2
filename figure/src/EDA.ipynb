```python
from tableone import TableOne

continuous_vars = [col for col in data.columns if col != 'label']
categorical_vars = []

data['label'] = data['label'] = data['label'].map({0: 'Jump', 1: 'Spin', 2: 'Step'})

# TableOne 생성
mytable = TableOne(data,
                  columns=continuous_vars,
                  categorical=categorical_vars,
                  groupby='label',
                  decimals=3)

# 결과 출력
print(mytable.tabulate(tablefmt="github"))
```

```python
import scipy.stats as stats
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Scientific notation setting
pd.set_option('display.float_format', lambda x: '%.2e' % x if abs(x) < 0.01 else '%.4f' % x)

features = [col for col in data.columns if col != 'label']
analysis_results = []

for feature in features:
    jump_data = data[data['label'] == 'Jump'][feature]
    spin_data = data[data['label'] == 'Spin'][feature]
    step_data = data[data['label'] == 'Step'][feature]

    f_stat, p_val = f_oneway(jump_data, spin_data, step_data)

    groups = [jump_data, spin_data, step_data]
    ss_between = sum(len(group) * ((group.mean() - data[feature].mean())**2) for group in groups)
    ss_total = sum((data[feature] - data[feature].mean())**2)
    eta_squared = ss_between / ss_total

    if eta_squared < 0.01:
        effect_size_interpretation = "Small Effect"
    elif eta_squared < 0.06:
        effect_size_interpretation = "Medium Effect"
    else:
        effect_size_interpretation = "Large Effect"

    desc_stats = data.groupby('label')[feature].agg(['mean', 'std']).round(4)

    analysis_results.append({
        'Feature': feature,
        'F-statistic': f_stat,
        'p-value': p_val,
        'Effect Size (η²)': eta_squared,
        'Effect Interpretation': effect_size_interpretation,
        'Jump_mean': desc_stats.loc['Jump', 'mean'],
        'Jump_std': desc_stats.loc['Jump', 'std'],
        'Spin_mean': desc_stats.loc['Spin', 'mean'],
        'Spin_std': desc_stats.loc['Spin', 'std'],
        'Step_mean': desc_stats.loc['Step', 'mean'],
        'Step_std': desc_stats.loc['Step', 'std']
    })

results_df = pd.DataFrame(analysis_results)
results_df = results_df.sort_values('Effect Size (η²)', ascending=False)

print("\n=== Overall Variable Analysis Results ===")
print(results_df)
```

```python
# 동작별 상위 10개의 박스플롯

top_10_features = results_df.head(10)['Feature'].tolist()

plt.figure(figsize=(15, 20))
for idx, feature in enumerate(top_10_features, 1):
    plt.subplot(5, 2, idx)
    sns.boxplot(x='label', y=feature, data=data)
    result_row = results_df[results_df['Feature']==feature].iloc[0]
    plt.title(f'{feature}\nη² = {result_row["Effect Size (η²)"]:.4f}\n{result_row["Effect Interpretation"]}')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.hist(results_df['Effect Size (η²)'], bins=30)
plt.axvline(x=0.01, color='r', linestyle='--', label='Small Effect Threshold')
plt.axvline(x=0.06, color='g', linestyle='--', label='Medium Effect Threshold')
plt.axvline(x=0.14, color='b', linestyle='--', label='Large Effect Threshold')
plt.title('Distribution of Effect Sizes (η²)')
plt.xlabel('Effect Size (η²)')
plt.ylabel('Count')
plt.legend()
plt.show()

```

```python
effect_size_summary = results_df['Effect Interpretation'].value_counts()
print("\n=== Effect Size Classification Summary ===")
print(effect_size_summary)

print("\n=== Top 20 Variables by Effect Size ===")
top_20 = results_df[['Feature', 'Effect Size (η²)', 'p-value', 'Effect Interpretation']].head(20)
print(top_20)

results_df['max_mean_diff'] = results_df.apply(lambda row: max([
    abs(row['Jump_mean'] - row['Spin_mean']),
    abs(row['Jump_mean'] - row['Step_mean']),
    abs(row['Spin_mean'] - row['Step_mean'])
]), axis=1)

print("\n=== Top 20 Variables by Mean Difference Between Groups ===")
mean_diff_top20 = results_df.sort_values('max_mean_diff', ascending=False)[
    ['Feature', 'max_mean_diff', 'Effect Size (η²)', 'Effect Interpretation', 'p-value']
].head(20)
print(mean_diff_top20)

results_df.to_csv('group_difference_analysis_results.csv', index=False)
```
